```python
import tensorflow as tf
import numpy as np

```

```python
xy = np.loadtxt("data/train_xor.txt", unpack=True)
x_data = xy[0:-1]
y_data = xy[-1]

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
W = tf.Variable(tf.random_uniform([1, len(x_data)], -1.0, 1.0))

h = tf.matmul(W, X)
h2 = tf.div(1., 1. + tf.exp(-h))

cost = -tf.reduce_mean(Y * tf.log(h2) + (1 - Y) * tf.log(1-h2))

opt = tf.train.GradientDescentOptimizer(0.01)
train = opt.minimize(cost)

init = tf.initialize_all_variables()
with tf.Session() as sess:
    sess.run(init)

    for step in range(2001):
        sess.run(train, feed_dict={X: x_data, Y: y_data})
        if step % 200 == 0:
            print (step, sess.run(cost, feed_dict={X: x_data, Y: y_data}), sess.run(W))
    
    co_predict = tf.equal(tf.floor(h + 0.5), Y)
    
    acc = tf.reduce_mean(tf.cast(co_predict, "float"))
    print(sess.run([h, tf.floor(h + 0.5), co_predict, acc], feed_dict={X: x_data, Y: y_data}))
    print("Acc:", acc.eval({X:x_data, Y: y_data}))
```
